```python
import numpy as np

def FtoC(vector):
    num = len(vector)
    trans_vector = []
    for i in range(num):
        target = vector[i]
        if target == 1:
            trans_vector.append(-3)
        elif target == 2:
            trans_vector.append(-2)
        elif target == 3:
            trans_vector.append(-1)
        elif target == 4:
            trans_vector.append(1)
        elif target == 5:
            trans_vector.append(2)
        elif target == 6:
            trans_vector.append(3)
        elif target == 7:
            trans_vector.append(4.5)
        elif target == 8:
            trans_vector.append(5)
        elif target == -2:
            trans_vector.append(-5)
        elif target == 0:
            trans_vector.append(0)
        else:
            print(target)
            print("没有被添加")
    return trans_vector

def CtoF(vector):
    num = len(vector)
    trans_vector = []
    for i in range(num):
        target = vector[i]
        if target == -3:
            trans_vector.append(1)
        elif target == -2:
            trans_vector.append(2)
        elif target == -1:
            trans_vector.append(3)
        elif target == 1:
            trans_vector.append(4)
        elif target == 2:
            trans_vector.append(5)
        elif target == 3:
            trans_vector.append(6)
        elif target == 4.5:
            trans_vector.append(7)
        elif target == 5:
            trans_vector.append(8)
        elif target == 0:
            trans_vector.append(0)
        elif target == -4:
            trans_vector.append(-1.5)
        elif target == -5:
            trans_vector.append(-2)
        elif target == -1.5:
            trans_vector.append(2.5)
        elif target == -6:
            trans_vector.append(-3)
        elif target == -7:
            trans_vector.append(-4)
        else:
            print(target)
            print("没有被添加")
    return trans_vector

def GtoC(vector):
    num = len(vector)
    trans_vector = []
    for i in range(num):
        target = vector[i]
        if target == 0:
            trans_vector.append(0)
        elif target == 1:
            trans_vector.append(-4)
        elif target == 2:
            trans_vector.append(-3)
        elif target == 3:
            trans_vector.append(-2)
        elif target == 5:
            trans_vector.append(1)
        elif target == -2:
            trans_vector.append(-6)
        elif target == -3:
            trans_vector.append(-7)
        else:
            print(target)
            print("没有被添加")
    return trans_vector

def CtoG(vector):
    num = len(vector)
    trans_vector = []
    for i in range(num):
        target = vector[i]
        if target == 0:
            trans_vector.append(0)
        elif target == -5:
            trans_vector.append(-1)
        elif target == -1:
            trans_vector.append(4.5)
        elif target == -4:
            trans_vector.append(1)
        elif target == -3:
            trans_vector.append(2)
        elif target == -2:
            trans_vector.append(3)
        elif target == 1:
            trans_vector.append(5)
        elif target == 2:
            trans_vector.append(6)
        elif target == 3:
            trans_vector.append(7)
        elif target == 5:
            trans_vector.append(9)
        elif target == -1.5:
            trans_vector.append(4)
        elif target == -6:
            trans_vector.append(-2)
        elif target == -7:
            trans_vector.append(-3)
        else:
            print(target)
            print("没有被添加")
    return trans_vector

def count_number(arr):
    result = {}
    for i in set(arr):
        result[i] = arr.count(i)
    return result

FtoC_list = FtoC([6,6, 6,6, 6,6, 5,3, 1,1,2,2,3,3,6,6])
print(FtoC_list)

GtoC_list = GtoC([2,2,2,2,2,2,1,2])

major_list = FtoC_list + GtoC_list
print(major_list)
count_number(major_list)

#映射顺序 2,3,-1,-2,-3,-4
def major_reflect(vector):
    num = len(vector)
    trans_vector = []
    for i in range(num):
        target = vector[i]
        if target == 2:
            trans_vector.append(0)
        elif target == 3:
            trans_vector.append(1)
        elif target == -1:
            trans_vector.append(2)
        elif target == -2:
            trans_vector.append(3)
        elif target == -3:
            trans_vector.append(4)
        elif target == -4:
            trans_vector.append(5)
        else:
            print(target)
            print("没有被添加")
    return trans_vector

reflect_major_list = major_reflect(major_list)
print(reflect_major_list)
len(reflect_major_list)

def markoving(darray, vector):
    num = len(vector)
    for i in range(num-1):   #最后一个不用考虑
        row = vector[i]
        column = vector[i+1]
        darray[row][column] += 1
    return darray

F1 = reflect_major_list[0:8]
F2 = reflect_major_list[8:16]
G1 = reflect_major_list[16:24]

major_array = np.zeros((6,6))

major_array = markoving(major_array,F1)
major_array = markoving(major_array,F2)
major_array = markoving(major_array,G1)

print(major_array)

def normalize_matrix(matrix):
    mat_size = matrix.shape[0]
    init_array = np.zeros((mat_size,mat_size))
    sum_row_list=[]
    for i in range(mat_size):
        count=0
        for j in range(mat_size):
            count+=matrix[i][j]
        sum_row_list.append(count)
    for k in range(mat_size):
        if sum_row_list[k] == 0:
            init_array[k][k] = 0
        else:
            init_array[k][k] = 1/sum_row_list[k]
    result = np.dot(init_array,matrix)
    
    return result

def sum_rows(matrix):
    sum_row_list1=[]
    for ii in range(len(matrix)):
        count=0
        for jj in range(len(matrix[ii])):
            count+=matrix[ii][jj]
        sum_row_list1.append(count)
    print(sum_row_list1)

markov_major = normalize_matrix(major_array)
print(markov_major)
sum_rows(markov_major)

#邻接矩阵

def generate_adj(darray, vector):
    num = len(vector)
    for i in range(num-1):
        row = vector[i]
        column = vector[i+1]
        darray[row][column] = 1
    return darray

adj_major = np.zeros((6,6))

adj_major = generate_adj(adj_major,F1)
adj_major = generate_adj(adj_major,F2)
adj_major = generate_adj(adj_major,G1)
print(adj_major)

#邻接矩阵调参数
k=7
x = 0.8

i = 0
Iter_adj_major = 1
for i in range(k):
    Iter_adj_major = np.dot(adj_major,Iter_adj_major)
    
Iter_adj_major = normalize_matrix(Iter_adj_major)

markov_major_final = x*markov_major + (1-x)*Iter_adj_major
sum_rows(markov_major_final)
print(markov_major_final)

#开始进行迭代
init_array = np.array([1/6,1/6,1/6,1/6,1/6,1/6]).reshape((-1,1))   #变成列向量
restmp = init_array

i = 0
for i in range(50):
    res = np.dot(markov_major_final.T,restmp)
    print (i, "\t", res)
    restmp = res

major_list = [2,3,-1,-2,-3,-4]

p_list1 = []

i = 0
for i in range(len(major_list)):
    target_list = res[i]
    p_list1.append(target_list[0])
print(p_list1)

major_12 = np.random.choice(major_list, 12, p=p_list1)

major1 = CtoF(major_12[0:4])

major2 = CtoG(major_12[4:8])

major3 = CtoF(major_12[8:12])

#小调分割线

def EmtoC(vector):
    num = len(vector)
    trans_vector = []
    for i in range(num):
        target = vector[i]
        if target == 0:
            trans_vector.append(0)
        elif target == 1:
            trans_vector.append(-2.5)
        elif target == 2:
            trans_vector.append(-1.5)
        elif target == 3:
            trans_vector.append(1)
        elif target == 5:
            trans_vector.append(2.5)
        elif target == -3:
            trans_vector.append(-5.5)
        else:
            print(target)
            print("没有被添加")
    return trans_vector

def CtoEm(vector):
    num = len(vector)
    trans_vector = []
    for i in range(num):
        target = vector[i]
        if target == 0:
            trans_vector.append(0)
        elif target == -2.5:
            trans_vector.append(1)
        elif target == -8.5:
            trans_vector.append(-6)
        elif target == -7:
            trans_vector.append(-5)
        elif target == -1.5:
            trans_vector.append(2)
        elif target == 1:
            trans_vector.append(3)
        elif target == -4:
            trans_vector.append(-2)
        elif target == -3:
            trans_vector.append(-1)
        elif target == -5.5:
            trans_vector.append(-3)
        elif target == 2.5:
            trans_vector.append(5)
        else:
            print(target)
            print("没有被添加")
    return trans_vector


def AmtoC(vector):
    num = len(vector)
    trans_vector = []
    for i in range(num):
        target = vector[i]
        if target == 1:
            trans_vector.append(-5.5)
        elif target == 2:
            trans_vector.append(-4)
        elif target == 3:
            trans_vector.append(-3)
        elif target == 4:
            trans_vector.append(-2.5)
        elif target == 5:
            trans_vector.append(-1.5)
        elif target == 6:
            trans_vector.append(1)
        elif target == 7:
            trans_vector.append(2)
        elif target == -2:
            trans_vector.append(-7)
        elif target == -3:
            trans_vector.append(-8.5)
        elif target == 0:
            trans_vector.append(0)
        else:
            print(target)
            print("没有被添加")
    return trans_vector

def CtoAm(vector):
    num = len(vector)
    trans_vector = []
    for i in range(num):
        target = vector[i]
        if target == -5.5:
            trans_vector.append(1)
        elif target == -8.5:
            trans_vector.append(-3)
        elif target == -7:
            trans_vector.append(-2)
        elif target == -4:
            trans_vector.append(2)
        elif target == -3:
            trans_vector.append(3)
        elif target == -2.5:
            trans_vector.append(4)
        elif target == 2.5:
            trans_vector.append(8)
        elif target == -1.5:
            trans_vector.append(5)
        elif target == 1:
            trans_vector.append(6)
        elif target == 2:
            trans_vector.append(7)
        elif target == 0:
            trans_vector.append(0)
        else:
            print(target)
            print("没有被添加")
    return trans_vector

EmtoC_list = EmtoC([5,5,5,5,5,5, 5,2,  1,1,1,1,1,1, 1,2])
AmtoC_list = AmtoC([3,3,2,2,1,1,-2,-2])


minor_list = EmtoC_list+AmtoC_list
count_number(minor_list)

#映射顺序 2.5, -1.5,-2.5,-3,-4,-5.5,-7
def minor_reflect(vector):
    num = len(vector)
    trans_vector = []
    for i in range(num):
        target = vector[i]
        if target == 2.5:
            trans_vector.append(0)
        elif target == -1.5:
            trans_vector.append(1)
        elif target == -2.5:
            trans_vector.append(2)
        elif target == -3:
            trans_vector.append(3)
        elif target == -4:
            trans_vector.append(4)
        elif target == -5.5:
            trans_vector.append(5)
        elif target == -7:
            trans_vector.append(6)
        else:
            print(target)
            print("没有被添加")
    return trans_vector


reflect_minor_list = minor_reflect(minor_list)
print(reflect_minor_list)
len(reflect_minor_list)

minor_array = np.zeros((7,7))

Em1 = reflect_minor_list[0:16]
Am1 = reflect_minor_list[16:24]


minor_array = markoving(minor_array,Em1)
minor_array = markoving(minor_array,Am1)
print(minor_array)

markov_minor = normalize_matrix(minor_array)
print(markov_minor)
sum_rows(markov_minor)

#第7行为空行，调一下参数

modify_line = np.array([1/7,1/7,1/7,1/7,1/7,1/7,1/7])
line7 = markov_minor[[6],:] 
lin7 = 0.8*line7 + 0.2*modify_line
markov_minor[[6],:]  = lin7
print(markov_minor)
sum_rows(markov_minor)

#邻接矩阵
adj_minor = np.zeros((7,7))

adj_minor = generate_adj(adj_minor,Em1)
adj_minor = generate_adj(adj_minor,Am1)

print(adj_minor)

k=7
x = 0.8

i = 0
Iter_adj_minor = 1
for i in range(k):
    Iter_adj_minor = np.dot(adj_minor,Iter_adj_minor)

print(Iter_adj_minor)
Iter_adj_minor = normalize_matrix(Iter_adj_minor)


#第五行为空行，调一下参数

modify_line = np.array([1/7,1/7,1/7,1/7,1/7,1/7,1/7])
Iter_adj_minor[[4],:] = modify_line
print(markov_minor)
sum_rows(Iter_adj_minor)



markov_minor_final = x*markov_minor + (1-x)*Iter_adj_minor
print(markov_minor_final)
sum_rows(markov_minor_final)

#开始进行迭代
init_array = np.array([1/7,1/7,1/7,1/7,1/7,1/7,1/7]).reshape((-1,1))   #变成列向量
restmp = init_array

i = 0
for i in range(50):
    res = np.dot(markov_minor_final.T,restmp)
    print (i, "\t", res)
    restmp = res

#映射顺序 2.5, -1.5,-2.5,-3,-4,-5.5,-7
minor_list = [2.5, -1.5,-2.5,-3,-4,-5.5,-7]
p_list2 = []

i = 0
for i in range(len(minor_list)):
    target_list = res[i]
    p_list2.append(target_list[0])
print(p_list2)


minor_12 = np.random.choice(minor_list, 12, p=p_list2)

minor1 = CtoEm(minor_12[0:8])
minor2 = CtoAm(minor_12[8:12])

solo_seq = major1+major2+minor1+minor2+major3 + [5,5,5,5,5,5,5,5]
print(solo_seq)
len(solo_seq)

def find_nearest(num, vector):
    result = num
    mark = 100
    length = len(vector)
    for i in range(length):
        instance = abs(num - vector[i])
        if mark > instance:
            mark = instance
            result = vector[i]
        else:
            continue
    return result

def post_process(vector):
    num = len(vector)
    five_list = [1,2,3,5,6, 8,9,10,12,13, -2,-3,-5,-6,-7,-9]
    C_list = [1,3,5,8,10,12,-7,-5,-3]
    Am_list = [1,3,6,8,10,-2,-5,-7]
    Em_list = [3,5,10,12,-3,-5]
    F_list = [1,6,8,13,-2,-7]
    G_list = [2,5,9,12,-3,-6]
    for i in range(num):
        target = vector[i]
#         int_part = math.floor(target/7)
#         remain_part = target - int_part*7
        if target == 0:
            vector[i] = vector[i-1]
            target = vector[i-1]
        if target in  five_list:
            other = 1
        else:
            if (i>=0 & i<4)|(i>=20 & i<24):
                vector[i] = find_nearest(vector[i],F_list)
            elif (i>=4 & i<8)|(i>=24 & i<32):
                #G chord
                vector[i] = find_nearest(vector[i],G_list)
            elif (i>=8 & i<16):
                #Em chord
                vector[i] = find_nearest(vector[i],Em_list)
            elif (i>=16 & i<20):
                #Am chord
                vector[i] = find_nearest(vector[i],Am_list)
                            
#     #律动感
#     vector[3] = vector[4] = vector[5] = vector[2]
#     vector[11] = vector[12] = vector[13] = vector[10]
#     vector[19] = vector[20] = vector[21] = vector[18]
#     vector[29] = vector[28] = vector[27] = vector[26]
    return vector

postprocessing_solo = post_process(solo_seq)
print(postprocessing_solo)





import mido
from mido import Message, MidiFile, MidiTrack

import math

def map_remain(num):
    if num == 1:
        result = 1
    elif num == 1.5:
        result = 2
    elif num == 2:
        result = 3
    elif num == 2.5:
        result = 4
    elif num == 3:
        result = 5
    elif num == 4:
        result = 6
    elif num == 4.5:
        result = 7
    elif num == 5:
        result = 8
    elif num == 5.5:
        result = 9
    elif num == 6:
        result = 10
    elif num == 6.5:
        result = 11
    elif num == 0:
        result = 0
    return result   


def map2_remain(num):
    if num == 1:
        result = 1
    elif num == 1.5:
        result = 2
    elif num == 2:
        result = 3
    elif num == 2.5:
        result = 4
    elif num == 3:
        result = 5
    elif num == 3.5:
        result = 6
    elif num == 4:
        result = 7
    elif num == 5:
        result = 8
    elif num == 5.5:
        result = 9
    elif num == 6:
        result = 10
    elif num == 6.5:
        result = 11
    elif num == 0:
        result = 0
    return result 
        
        
def toMidi(vector):
    
    num = len(vector)
    trans_vector = []
    for i in range(num):
        target = vector[i]
        if target > 0:
            base = 59   #基数是59
            int_part = math.floor(target)     #得到target的整数部分
            times = math.floor(int_part/7)    #整数部分是由多少个7组成的
            base += times*12

            remain = target - times*7        #得到剩余部分
            base += map_remain(remain)
            trans_vector.append(base)
        else:  #target < 0
            base = 60   #基数是60
            target = abs(target)
            int_part = math.floor(target)
            times = math.floor(int_part/7)
            base -= times*12
            
            remain = target - times*7
            base -= map2_remain(remain)
            trans_vector.append(base)
    return trans_vector

mid = MidiFile()       #给自己的文件定的.mid后缀
track = MidiTrack()    #定义声部，一个MidoTrack()就是一个声部

def beat(time):            #与mido的拍子互换  
    time /= 60 * 1000  
    time = 1/time  
    return time  

def yin(yin,pai,qian=0,unit = track, liang=64,qi=2): #yin是音，pai是指时间（节拍），
    if type(yin)== str:               #qian是是音前多少拍
        yin = num(yin)                #liang是音量

    unit.append(Message('program_change',channel=0,program=qi,time=0))
    unit.append(Message('note_on',note=yin,velocity=liang,time=qian,channel=0))
    unit.append(Message('note_off',note=yin,velocity=liang,time=int(beat(pai)),channel=0))

midi_seq = toMidi(postprocessing_solo)
for i in range(len(postprocessing_solo)):
    
    yin(midi_seq[i],80, 0,track,80,2)   
    

track1 = MidiTrack()
track2 = MidiTrack()
track3 = MidiTrack()

def F_chord(track1,track2,track3):
    sq = toMidi([-4,-2,1])
    yin(sq[0],20, 0,track1,80,2)
    yin(sq[1],20, 0,track2,80,2)
    yin(sq[2],20, 0,track3,80,2)
    
def C_chord(track1,track2,track3):
    C_sq = toMidi([1,3,5])
    yin(C_sq[0],20, 0,track1,80,2)
    yin(C_sq[1],20, 0,track2,80,2)
    yin(C_sq[2],20, 0,track3,80,2)
    
def G_chord(track1,track2,track3):
    sq = toMidi([-3,-1,2])
    yin(sq[0],20, 0,track1,80,2)
    yin(sq[1],20, 0,track2,80,2)
    yin(sq[2],20, 0,track3,80,2)
    
def Em_chord(track1,track2,track3):
    sq = toMidi([-5,-3,-1])
    yin(sq[0],20, 0,track1,80,2)
    yin(sq[1],20, 0,track2,80,2)
    yin(sq[2],20, 0,track3,80,2)

def Am_chord(track1,track2,track3):
    sq = toMidi([-2,1,3])
    yin(sq[0],20, 0,track1,80,2)
    yin(sq[1],20, 0,track2,80,2)
    yin(sq[2],20, 0,track3,80,2)
    

F_chord(track1,track2,track3)
G_chord(track1,track2,track3)
Em_chord(track1,track2,track3)
Em_chord(track1,track2,track3)

Am_chord(track1,track2,track3)
F_chord(track1,track2,track3)
G_chord(track1,track2,track3)
G_chord(track1,track2,track3)

mid.tracks.append(track)
mid.tracks.append(track1)
mid.tracks.append(track2)
mid.tracks.append(track3)

mid.save('C:\\Users\\Administrator\\Desktop\\final\\solo\\SOLOK=7.mid')
```
