```python
def generate_C_solo(solo_seq,solo_prob):
    seq1 = np.random.choice(solo_seq, 12, p=solo_prob)
    seq1 = seq1.tolist()
    #添加后缀八个琶音
    choiceArr = np.random.choice([1,2,3,4],1,p=[1/4,1/4,1/4,1/4])     #四种可选情况，随机选择
    
    if choiceArr[0] == 1:
        seq2 = [1,2,4,4.5, 5,5,5,5]
    elif choiceArr[0] == 2:
        seq2 = [5,4.5,4,2.5, 1,1,1,1]
        
    else:
        seq2 = np.random.choice(solo_seq, 3, p=solo_prob)
        seq2 = seq2.tolist()
        if choiceArr[0] == 3:
            seq2 += [4,4.5,5,6.5,7]
        elif choiceArr[0] == 4:
            seq2 += [7,6.5,5,4.5,4]
            

    seq1[6] = seq1[5]
    seq1[7] = seq1[5]
    return seq1+seq2
```

```python
import math

def map_remain(num):
    if num == 1:
        result = 1
    elif num == 1.5:
        result = 2
    elif num == 2:
        result = 3
    elif num == 2.5:
        result = 4
    elif num == 3:
        result = 5
    elif num == 4:
        result = 6
    elif num == 4.5:
        result = 7
    elif num == 5:
        result = 8
    elif num == 5.5:
        result = 9
    elif num == 6:
        result = 10
    elif num == 6.5:
        result = 11
    elif num == 0:
        result = 0
    return result   


def map2_remain(num):
    if num == 1:
        result = 1
    elif num == 1.5:
        result = 2
    elif num == 2:
        result = 3
    elif num == 2.5:
        result = 4
    elif num == 3:
        result = 5
    elif num == 3.5:
        result = 6
    elif num == 4:
        result = 7
    elif num == 5:
        result = 8
    elif num == 5.5:
        result = 9
    elif num == 6:
        result = 10
    elif num == 6.5:
        result = 11
    elif num == 0:
        result = 0
    return result 
        
        
def toMidi(vector):
    
    num = len(vector)
    trans_vector = []
    for i in range(num):
        target = vector[i]
        if target > 0:
            base = 59   #基数是59
            int_part = math.floor(target)     #得到target的整数部分
            times = math.floor(int_part/7)    #整数部分是由多少个7组成的
            base += times*12

            remain = target - times*7        #得到剩余部分
            base += map_remain(remain)
            trans_vector.append(base)
#         elif target == 0:
#             last = trans_vector[len(trans_vector)]
#             trans_vector.append(last)
        elif target == 0:
            trans_vector.append(0)
        else:  #target < 0
            base = 60   #基数是60
            target = abs(target)
            int_part = math.floor(target)
            times = math.floor(int_part/7)
            base -= times*12
            
            remain = target - times*7
            base -= map2_remain(remain)
            trans_vector.append(base)
    return trans_vector
```

```python
mid = MidiFile()       #给自己的文件定的.mid后缀
track = MidiTrack()    #定义声部，一个MidoTrack()就是一个声部

def beat(time):            #与mido的拍子互换  
    time /= 60 * 1000  
    time = 1/time  
    return time  

def yin(yin,pai,qian=0,unit = track, liang=64,qi=2): #yin是音，pai是指时间（节拍），
    if type(yin)== str:               #qian是是音前多少拍
        yin = num(yin)                #liang是音量

    unit.append(Message('program_change',channel=0,program=qi,time=0))
    unit.append(Message('note_on',note=yin,velocity=liang,time=qian,channel=0))
    unit.append(Message('note_off',note=yin,velocity=liang,time=int(beat(pai)),channel=0))
    
def find_nearest(num, vector):
    result = num
    mark = 100
    length = len(vector)
    for i in range(length):
        instance = abs(num - vector[i])
        if mark > instance:
            mark = instance
            result = vector[i]
        else:
            continue
    if num == 0:
        result = 0
    return result
```
