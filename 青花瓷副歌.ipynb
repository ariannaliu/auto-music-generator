```python
def normalize_matrix(matrix):
    mat_size = matrix.shape[0]
    init_array = np.zeros((mat_size,mat_size))
    sum_row_list=[]
    for i in range(mat_size):
        count=0
        for j in range(mat_size):
            count+=matrix[i][j]
        sum_row_list.append(count)
    for k in range(mat_size):
        if sum_row_list[k] == 0:
            init_array[k][k] = 0
        else:
            init_array[k][k] = 1/sum_row_list[k]
    result = np.dot(init_array,matrix)
    
    return result

def sum_rows(matrix):
    sum_row_list1=[]
    for ii in range(len(matrix)):
        count=0
        for jj in range(len(matrix[ii])):
            count+=matrix[ii][jj]
        sum_row_list1.append(count)
    print(sum_row_list1)
```

```python
def find_nearest(num, vector):
    result = num
    mark = 100
    length = len(vector)
    for i in range(length):
        instance = abs(num - vector[i])
        if mark > instance:
            mark = instance
            result = vector[i]
        else:
            continue
    return result

def post_process(vector):
    num = len(vector)
    five_list = [1,2,3,5,6, 8,9,10,12,13, -2,-3,-5,-6,-7,-9]
    C_list = [1,3,5,8,10,12,-7,-5,-3]
    Am_list = [1,3,6,8,10,-2,-5,-7]
    Em_list = [3,5,10,12,-3,-5]
    F_list = [1,6,8,13,-2,-7]
    G_list = [2,5,9,12,-3,-6]
    for i in range(num):
        target = vector[i]
#         int_part = math.floor(target/7)
#         remain_part = target - int_part*7
        if target == 0:
            vector[i] = vector[i-1]
            target = vector[i-1]
        if target in  five_list:
            other = 1
        else:
            if (i>=0 & i<4)|(i>=16 & i<20)|(i>=32 & i<36)|(i>=48 & i<52):
                vector[i] = find_nearest(vector[i],F_list)
            elif (i>=4 & i<8)|(i>=20 & i<24)|(i>=36 & i<40)|(i>=52 & i<56):
                #G chord
                vector[i] = find_nearest(vector[i],G_list)
            elif (i>=8 & i<12)|(i>=40 & i<44):
                #Em chord
                vector[i] = find_nearest(vector[i],Em_list)
            elif (i>=12 & i<16)|(i>=44 & i<48):
                #Am chord
                vector[i] = find_nearest(vector[i],Am_list)
            elif (i>=24 & i<32)|(i>=56 & i<64):
                #C chord
                vector[i] = find_nearest(vector[i],C_list)
                            
    #律动感
    vector[13] = vector[12]
    vector[29] = vector[28]
    vector[45] = vector[44]
    return vector
```

```python
import math

def map_remain(num):
    if num == 1:
        result = 1
    elif num == 1.5:
        result = 2
    elif num == 2:
        result = 3
    elif num == 2.5:
        result = 4
    elif num == 3:
        result = 5
    elif num == 4:
        result = 6
    elif num == 4.5:
        result = 7
    elif num == 5:
        result = 8
    elif num == 5.5:
        result = 9
    elif num == 6:
        result = 10
    elif num == 6.5:
        result = 11
    elif num == 0:
        result = 0
    return result   


def map2_remain(num):
    if num == 1:
        result = 1
    elif num == 1.5:
        result = 2
    elif num == 2:
        result = 3
    elif num == 2.5:
        result = 4
    elif num == 3:
        result = 5
    elif num == 3.5:
        result = 6
    elif num == 4:
        result = 7
    elif num == 5:
        result = 8
    elif num == 5.5:
        result = 9
    elif num == 6:
        result = 10
    elif num == 6.5:
        result = 11
    elif num == 0:
        result = 0
    return result 
        
        
def toMidi(vector):
    
    num = len(vector)
    trans_vector = []
    for i in range(num):
        target = vector[i]
        if target > 0:
            base = 59   #基数是59
            int_part = math.floor(target)     #得到target的整数部分
            times = math.floor(int_part/7)    #整数部分是由多少个7组成的
            base += times*12

            remain = target - times*7        #得到剩余部分
            base += map_remain(remain)
            trans_vector.append(base)
        else:  #target < 0
            base = 60   #基数是60
            target = abs(target)
            int_part = math.floor(target)
            times = math.floor(int_part/7)
            base -= times*12
            
            remain = target - times*7
            base -= map2_remain(remain)
            trans_vector.append(base)
    return trans_vector
```
