```python
def FtoC(vector):
    num = len(vector)
    trans_vector = []
    for i in range(num):
        target = vector[i]
        if target == 1:
            trans_vector.append(-3)
        elif target == 2:
            trans_vector.append(-2)
        elif target == 3:
            trans_vector.append(-1)
        elif target == 4:
            trans_vector.append(1)
        elif target == 5:
            trans_vector.append(2)
        elif target == 6:
            trans_vector.append(3)
        elif target == 7:
            trans_vector.append(4.5)
        elif target == 8:
            trans_vector.append(5)
        elif target == 0:
            trans_vector.append(0)
    print(trans_vector)
    return trans_vector

def CtoF(vector):
    num = len(vector)
    trans_vector = []
    for i in range(num):
        target = vector[i]
        if target == -3:
            trans_vector.append(1)
        elif target == -2:
            trans_vector.append(2)
        elif target == -1:
            trans_vector.append(3)
        elif target == 1:
            trans_vector.append(4)
        elif target == 2:
            trans_vector.append(5)
        elif target == 3:
            trans_vector.append(6)
        elif target == 4.5:
            trans_vector.append(7)
        elif target == 5:
            trans_vector.append(8)
        elif target == 0:
            trans_vector.append(0)
        elif target == -4:
            trans_vector.append(-1.5)
        elif target == -1.5:
            trans_vector.append(2.5)
        elif target == -6:
            trans_vector.append(-3)
    return trans_vector
```

```python
def GtoC(vector):
    num = len(vector)
    trans_vector = []
    for i in range(num):
        target = vector[i]
        if target == 1:
            trans_vector.append(-4)
        elif target == 2:
            trans_vector.append(-3)
        elif target == 3:
            trans_vector.append(-2)
        elif target == 4:
            trans_vector.append(-1.5)
        elif target == 5:
            trans_vector.append(1)
        elif target == 6:
            trans_vector.append(2)
        elif target == 7:
            trans_vector.append(3)
        elif target == -2:
            trans_vector.append(-6)
        elif target == 0:
            trans_vector.append(0)
    print(trans_vector)
    return trans_vector

def CtoG(vector):
    num = len(vector)
    trans_vector = []
    for i in range(num):
        target = vector[i]
        if target == -4:
            trans_vector.append(1)
        elif target == -3:
            trans_vector.append(2)
        elif target == -2:
            trans_vector.append(3)
        elif target == -1.5:
            trans_vector.append(4)
        elif target == 1:
            trans_vector.append(5)
        elif target == 2:
            trans_vector.append(6)
        elif target == 3:
            trans_vector.append(7)
        elif target == -6:
            trans_vector.append(-2)
        elif target == 0:
            trans_vector.append(0)
        elif target == -1:
            trans_vector.append(4.5)
        elif target == 4.5:
            trans_vector.append(8.5)
        elif target == 5:
            trans_vector.append(9)
    return trans_vector
```

```python
def AmtoC(vector):
    num = len(vector)
    trans_vector = []
    for i in range(num):
        target = vector[i]
        if target == 1:
            trans_vector.append(-5.5)
        elif target == 2:
            trans_vector.append(-4)
        elif target == 3:
            trans_vector.append(-3)
        elif target == 4:
            trans_vector.append(-2.5)
        elif target == 5:
            trans_vector.append(-1.5)
        elif target == 6:
            trans_vector.append(1)
        elif target == 7:
            trans_vector.append(2)
        elif target == -2:
            trans_vector.append(-7)
        elif target == 0:
            trans_vector.append(0)
    print(trans_vector)
    return trans_vector

def CtoAm(vector):
    num = len(vector)
    trans_vector = []
    for i in range(num):
        target = vector[i]
        if target == -5.5:
            trans_vector.append(1)
        elif target == -4:
            trans_vector.append(2)
        elif target == -3:
            trans_vector.append(3)
        elif target == -2.5:
            trans_vector.append(4)
        elif target == -1.5:
            trans_vector.append(5)
        elif target == 1:
            trans_vector.append(6)
        elif target == 2:
            trans_vector.append(7)
        elif target == -7:
            trans_vector.append(-2)
        elif target == 0:
            trans_vector.append(0)
    return trans_vector
```

```python
def count_number(arr):
    result = {}
    for i in set(arr):
        result[i] = arr.count(i)
    return result
```

```python
def reflect_major(vector):
    num = len(vector)
    trans_vector = []
    for i in range(num):
        target = vector[i]
        if target == 0:
            trans_vector.append(0)
        elif target == 1:
            trans_vector.append(1)
        elif target == 2:
            trans_vector.append(2)
        elif target == 3:
            trans_vector.append(3)
        elif target == 4.5:
            trans_vector.append(4)
        elif target == -1:
            trans_vector.append(5)
        elif target == -2:
            trans_vector.append(6)
        elif target == -3:
            trans_vector.append(7)
        elif target == -4:
            trans_vector.append(8)
    print(trans_vector)
    return trans_vector

def reflect_minor(vector):
    num = len(vector)
    trans_vector = []
    for i in range(num):
        target = vector[i]
        if target == 0:
            trans_vector.append(0)
        elif target == 1:
            trans_vector.append(1)
        elif target == 2:
            trans_vector.append(2)
        elif target == -1.5:
            trans_vector.append(3)
        elif target == -3:
            trans_vector.append(4)
        elif target == -4:
            trans_vector.append(5)
        elif target == -5.5:
            trans_vector.append(6)
    print(trans_vector)
    return trans_vector
```
